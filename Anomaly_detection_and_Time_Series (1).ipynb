### **1. What is Dimensionality Reduction? Why is it important in machine learning?**
**Answer**:

Dimensionality Reduction is the process of reducing the number of input variables in a dataset while retaining as much relevant information as possible. It transforms high-dimensional data into a lower-dimensional form.

**Importance**:

Avoids the curse of dimensionality and improves model generalization.

Reduces overfitting by eliminating redundant or irrelevant features.

Enhances computational efficiency and model interpretability.

Helps in visualizing complex data in 2D or 3D space.

### **2. Name and briefly describe three common dimensionality reduction techniques.**
**Answer**:

**1. Principal Component Analysis (PCA)**:
A linear technique that transforms correlated features into uncorrelated variables called principal components, capturing maximum variance.

**2. Linear Discriminant Analysis (LDA)**:
A supervised technique that projects data to maximize class separability by focusing on the directions that best distinguish different classes.

**3. t-Distributed Stochastic Neighbor Embedding (t-SNE)**:
A non-linear method that reduces dimensions while preserving local structure, commonly used for visualizing high-dimensional data.

### **4. Explain the concept of anomaly detection and its significance.**

**Answer**:

Anomaly Detection is the process of identifying data points that deviate significantly from the majority of data. These “outliers” often indicate critical events such as fraud, system failures, or data errors.

**Significance**:

* Detects unusual patterns in financial transactions, network security, or industrial systems.

* Improves system reliability and safety.

* Helps in quality control and fault detection.

### **5. List and briefly describe three types of anomaly detection techniques.**

**Answer**:

**1. Statistical Methods**:

Detect anomalies by modeling normal data distribution (e.g., z-score, Gaussian models).

**2. Machine Learning–Based Methods**:

Use algorithms like Isolation Forest or One-Class SVM to learn the boundary of normal data.

**3. Distance/Clustering-Based Methods**:

Identify anomalies based on their distance from cluster centers or nearest neighbors (e.g., KNN-based outlier detection).

### **6. What is time series analysis? Mention two key components of time series data**.

**Answer**:

Time Series Analysis involves studying data points collected over time to identify patterns, trends, and dependencies for forecasting.

**Key Components**:

**1. Trend**: The long-term direction of data movement (upward, downward, or constant).

**2. Seasonality**: Regular, repeating patterns that occur at fixed intervals (e.g., monthly, quarterly).

### **7. Describe the difference between seasonality and cyclic behavior in time series**.

**Answer**:

* **Seasonality**: Refers to repetitive patterns occurring at fixed and known intervals (e.g., daily, monthly). Example: increased retail sales every December.

* **Cyclic Behavior**: Refers to irregular and non-fixed patterns influenced by economic or natural cycles (e.g., business cycles).

### **8. Write Python code to perform K-means clustering on a sample dataset**.

**Answer**:

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering Results")
plt.legend()
plt.show()
```

### **9. What is inheritance in OOP? Provide a simple example in Python**.

**Answer**:

Inheritance is an OOP concept that allows a class (child/subclass) to inherit attributes and methods from another class (parent/superclass), promoting code reuse and extensibility.

**Example**:

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

d = Dog()
print(d.speak())
```
